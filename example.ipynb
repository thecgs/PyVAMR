```python
#!/usr/bin/env python
# coding: utf-8

import re
import os
from Bio import SeqIO, Entrez

from Bio.SeqFeature import CompoundLocation, SimpleLocation

def get_genbank_from_ncbi(accession):
    Entrez.email = "thecgs001.foxmail.com"
    gb_text = Entrez.efetch(db="Nucleotide", id=accession, rettype='gb')
    return gb_text

accession = "NC_009093"

gb = get_genbank_from_ncbi(accession)
out = open(f"/mnt/nfs1/jupyter/VMTCP/example/{accession}.gbk", 'w')
for i in gb:
    print(i, end='', file=out)
out.close()
```
